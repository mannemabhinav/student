```python
import pandas as pd
df=pd.read_csv("/content/student.csv")
print(df)
```

```python
print(df.info())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
department_marks=df.groupby('Department')['Marks'].sum()
print(department_marks)
```

```python
import matplotlib.pyplot as plt
department_marks.plot(kind='bar')
plt.xlabel('Department')
plt.ylabel('Marks')
plt.title('Total marks by Depatment')
plt.show()
```

```python
dept_marks = df.groupby("Department")["Marks"].mean()

print("\nAverage Marks by Department:")
print(dept_marks)
```

```python
top_student = df.loc[df["Marks"].idxmax()]
print("\nHighest Scoring Student:")
print(top_student)
```

```python
print("\nOverall Average Marks:")
print(df["Marks"].mean())

```

```python
student_marks = df.groupby("Name")["Marks"].sum()
student_marks.plot(kind="bar")
plt.title("Marks by Student")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```
